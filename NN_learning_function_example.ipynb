```python
# import relevant libraries, numpy, pandas, tensorflow, datasets
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.models import Sequential
import math
import numpy as np
import random
from sklearn.metrics import mean_squared_error
```

```python
# create a model and instantiating layers

# NN model using convex functions
model = Sequential([
    Dense(units=16, input_dim=1, activation='LeakyReLU'),
    Dense(units=4, input_dim=1, activation='LeakyReLU'),
    Dense(units=1, input_dim=1, activation='LeakyReLU')
])

model.summary()
```

```python
# generate random data, arrays with numbers

# pairs ranging from [-2, 2]

np.random.seed(36)

X_train = np.random.uniform(-2, 2, 100) # array of 100 numbers in range of [-2, 2]

# results for (x^2 - 1)^2 to train model
Y_train = np.array([math.pow((x*x - 1), 2) for x in X_train])
```

```python
# compile model

model.compile(optimizer='adam', loss='mean_absolute_error')

# train model

model.fit(X_train, Y_train, epochs=200, batch_size=10, shuffle=True)
```

```python
# Generate new random data for prediction, to test models

X_test = np.random.uniform(-2, 2, 10)

predictions = model.predict(X_test).flatten()
```

```python
# test results:

y_test_actual = np.array([math.pow((x*x - 1), 2) for x in X_test])
```

```python
# Calculate mean squared error

mse = mean_squared_error(y_test_actual, predictions)
```

```python
# Checking test results:

# Print actual sums, predicted sums, and MSE

print("\nActual f(x)  |  Predicted f(x)")
print("---------------------------")
for i in range(len(y_test_actual)):
    print(f"{y_test_actual[i]:.3f}      |  {predictions[i]:.3f}")

print(f"\nmean_Squared_error: {mse:.3f}")
```
